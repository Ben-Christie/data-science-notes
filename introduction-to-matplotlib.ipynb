### Two Options for Subplots

```python
# subplot option 1 -> each ax is an individual plot, ax1 and ax2 are in brackets cause they're on the same line and the same for ax3 and ax4
# nrows and ncols just outlines the number of rows and columns
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows = 2,
    ncols = 2,
    figsize = (10, 5)
)

# set space between rows (hspace) -> I assume wspace would set with between cols
plt.subplots_adjust(hspace=0.5)

# plot to different ax
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nb_prices.keys(), nb_prices.values())
ax4.hist(np.random.randn(1000))

ax1.set(title='ax1')
ax2.set(title='ax2')
ax3.set(title='ax3')
ax4.set(title='ax4')

plt.show()
```

```python
# subplots option 2 (not recommended)
fig, ax = plt.subplots(
    nrows = 2,
    ncols = 2,
    figsize=(10, 5)
)

# set space between rows (hspace) -> I assume wspace would set with between cols
plt.subplots_adjust(hspace=0.5)

# plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nb_prices.keys(), nb_prices.values())
ax[1, 1].hist(np.random.randn(1000))

ax[0, 0].set(title='[0, 0]')
ax[0, 1].set(title='[0, 1]')
ax[1, 0].set(title='[1, 0]')
ax[1, 1].set(title='[1, 1]')

plt.show()
```

#### Object Oriented Approach to Plotting (recommended approach)

```python
heart_disease
```

```python
# create a subset of people over 50
over_50 = heart_disease[heart_disease['age'] >= 50]
over_50.head()
```

```python
# c =  colour
over_50.plot(kind='scatter', x='age', y='chol', c='target')
plt.show()
```

The above doesn't look great, this is where we use the OO method as it allows for more complexity

```python
# object-oriented method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6)) # <- width, height

# ax=ax tells matplotlib to plot on the ax that we created above
over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target',
    ax=ax
)

# ax = the plot, fig is the entire frame i.e. you can have multiple axes in a fig

# set the x-axis limit to be from 45 to 100
ax.set_xlim([45, 100])

plt.show()
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6)) # <- width, height

# plot the data -> c = colour
scatter = ax.scatter(
    x=over_50['age'], 
    y=over_50['chol'],
    c=over_50['target']
)

# customise the ax
ax.set(
    title='Age/Cholesterol Level Comparison',
    xlabel='Age',
    ylabel='Cholesterol Level'
)

# add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# add a horizontal line for the mean
ax.axhline(
    over_50['chol'].mean(),
    linestyle='--'
)

plt.show()
```

#### Comparing Cholesterol and Max Heart Rate

```python
heart_disease.head()
```

```python
# create the plot size
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(
    x=heart_disease['thalach'],
    y=heart_disease['chol'],
    c=heart_disease['target']
)

# customise plot
ax.set(
    title='Max Heart Rate/Cholesterol Level Comparison for Heart Disease Analysis',
    xlabel='Maximum Heart Rate',
    ylabel='Cholesterol Level'
)

# add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# add horizontal line for the chol mean
ax.axhline(
    heart_disease['chol'].mean(),
    linestyle='--'
)

plt.show()
```

```python
# Subplotting both Heart Rate and Cholesterol Levels

# sharex = share the x axis
fig, (ax1, ax2) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10, 10),
    sharex=True
)

# add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')

# add data to ax1
scatter = ax1.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target']
)

# customise ax1
ax1.set(
    title='Spread of Cholesterol Levels for Over 50\'s',
    ylabel='Cholesterol Level'
)

# add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# add horizontal line for mean cholesterol
ax1.axhline(
    over_50['chol'].mean(),
    linestyle='--'
)

# add data to ax2
scatter2 = ax2.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target']
)

# customise ax2
ax2.set(
    title='Spread of Heart Rates for Over 50\'s',
    xlabel='Age',
    ylabel='Heart Rate'
)

# add legend to ax2
ax2.legend(*scatter.legend_elements(), title='Target')

# add horizontal line for mean cholesterol
ax2.axhline(
    over_50['thalach'].mean(),
    linestyle='--'
)

plt.show()
```

```python
# set the style for a chart we've created from scratch
plt.style.use('seaborn-v0_8-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6)) # <- width, height

# plot the data -> c = colour, cmap = colour map, changes the colour scheme, checkout colors documentation on matplotlib
scatter = ax.scatter(
    x=over_50['age'], 
    y=over_50['chol'],
    c=over_50['target'],
    cmap='winter'
)

# customise the ax
ax.set(
    title='Age/Cholesterol Level Comparison',
    xlabel='Age',
    ylabel='Cholesterol Level'
)

# add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# add a horizontal line for the mean
ax.axhline(
    over_50['chol'].mean(),
    linestyle='--'
)

plt.show()
```

```python
# sharex = share the x axis
fig, (ax1, ax2) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10, 10),
    sharex=True
)

# add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')

# add data to ax1
scatter = ax1.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
    cmap='winter'
)

# customise ax1
ax1.set(
    title='Spread of Cholesterol Levels for Over 50\'s',
    ylabel='Cholesterol Level'
)

# set x-axis range
ax1.set_xlim([45, 80])

# add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# add horizontal line for mean cholesterol
ax1.axhline(
    over_50['chol'].mean(),
    linestyle='--'
)

# add data to ax2
scatter2 = ax2.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target'],
    cmap='spring'
)

# customise ax2
ax2.set(
    title='Spread of Heart Rates for Over 50\'s',
    xlabel='Age',
    ylabel='Heart Rate'
)

# set x-axis range
ax2.set_xlim([45, 80])
ax2.set_ylim([60, 200])

# add legend to ax2
ax2.legend(*scatter2.legend_elements(), title='Target')

# add horizontal line for mean cholesterol
ax2.axhline(
    over_50['thalach'].mean(),
    linestyle='--'
)

plt.show()
```
